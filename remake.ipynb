```python
aspect = pd.read_csv("full_aspect.csv")
og = pd.read_csv("../tripadvisor.csv")
```

```python
# 1. aspect:sentiment 형태 컬럼 만들기
aspect["aspect_sentiment"] = aspect["aspect"].astype(str) + ":" + aspect["sentiment"].astype(str)

# 2. reviewID별로 하나의 리스트로 묶기
aspect_grouped = (
    aspect.groupby("reviewID")["aspect_sentiment"]
    .apply(list)
    .reset_index()
)

# 3. og랑 merge
merged = og.merge(
    aspect_grouped,
    left_on="ReviewID",
    right_on="reviewID",
    how="left"
)

# 4. 필요 없으면 reviewID 컬럼 제거
merged = merged.drop(columns=["reviewID"])

merged.head()
```

```python
merged.columns
```

```python
tmp = merged[["location_name", "aspect_sentiment"]].explode("aspect_sentiment")

tmp[["aspect", "sentiment"]] = tmp["aspect_sentiment"].str.split(":", expand=True)

profile = (
    tmp.groupby(["location_name", "aspect", "sentiment"])
       .size()
       .reset_index(name="count")
)
```

```python
profile
```

```python
# Positive Top10
positive_top10 = (
    profile[profile["sentiment"] == "positive"]
    .sort_values(["location_name", "count"], ascending=[True, False])
    .groupby("location_name")
    .head(10)
)

# Negative Top10
negative_top10 = (
    profile[profile["sentiment"] == "negative"]
    .sort_values(["location_name", "count"], ascending=[True, False])
    .groupby("location_name")
    .head(10)
)
```

```python
positive_top10
```

```python
aspect_map = {
    "view": "views",
    "views": "views",
    "scenery": "views",
    "landscape": "views",
    "night_view": "views",
    "city_view": "views",

    "history": "history",
    "historical": "history",
    "culture": "culture",
    "tradition": "culture",

    "architecture": "architecture",
    "building": "architecture",
    "palace": "architecture",
    "temple": "architecture",

    "crowd": "crowds",
    "crowds": "crowds",
    "crowded": "crowds",
    "queue": "crowds",
    "waiting_time": "crowds",

    "transportation": "accessibility",
    "subway": "accessibility",
    "bus": "accessibility",
    "location": "accessibility",
    "access": "accessibility",

    "price": "price",
    "ticket": "price",
    "entrance_fee": "price",
    "cost": "price",

    "food": "food",
    "restaurant": "food",
    "cafe": "food",

    "shopping": "shopping",
    "shops": "shopping",
    "market": "shopping",

    "cleanliness": "cleanliness",
    "clean": "cleanliness",
    "toilet": "facilities",
    "restroom": "facilities",
    "facilities": "facilities",

    "guide": "service",
    "guides": "service",
    "staff": "service",
    "service": "service",

    "atmosphere": "atmosphere",
    "vibe": "atmosphere",
    "ambience": "atmosphere",

    "activities": "activities",
    "experience": "activities",
    "performance": "activities",
    "tour": "activities",
}
```

```python
profile["aspect_norm"] = (
    profile["aspect"]
    .str.lower()
    .str.strip()
    .map(aspect_map)
)

# 매핑 안 된 aspect는 일단 원래 aspect 유지
profile["aspect_norm"] = profile["aspect_norm"].fillna(profile["aspect"])
```

```python
profile_norm = (
    profile.groupby(["location_name", "aspect_norm", "sentiment"])["count"]
    .sum()
    .reset_index()
    .rename(columns={"aspect_norm": "aspect"})
)
```

```python
positive_top10 = (
    profile_norm[profile_norm["sentiment"] == "positive"]
    .sort_values(["location_name", "count"], ascending=[True, False])
    .groupby("location_name")
    .head(10)
)

negative_top10 = (
    profile_norm[profile_norm["sentiment"] == "negative"]
    .sort_values(["location_name", "count"], ascending=[True, False])
    .groupby("location_name")
    .head(10)
)
```

```python
positive_summary = (
    positive_top10
    .groupby("location_name")
    .apply(lambda x: ", ".join(f"{a}({c})" for a, c in zip(x["aspect"], x["count"])))
    .reset_index(name="positive_top10")
)

negative_summary = (
    negative_top10
    .groupby("location_name")
    .apply(lambda x: ", ".join(f"{a}({c})" for a, c in zip(x["aspect"], x["count"])))
    .reset_index(name="negative_top10")
)

tourist_profile = positive_summary.merge(
    negative_summary,
    on="location_name",
    how="outer"
)

tourist_profile.head()
```

```python
tourist_profile.to_csv("tourist_profile.csv", index=False, encoding="utf-8-sig")
```

```python
remove_locations = [
    "61. MJ Jeju Diving Club",
    "63. Seoul Free Walking Tour",
    "57. Seoul Pub Crawl",
    "41. Seoul Sky"
]

tourist_profile = tourist_profile[
    ~tourist_profile["location_name"].isin(remove_locations)
].reset_index(drop=True)
```

```python
# og 데이터에서 관광지별 평균 rating 계산
rating_mean = (
    og.groupby("location_name")["review_rating"]
    .mean()
    .reset_index()
    .rename(columns={"review_rating": "avg_rating"})
)

# tourist_profile에 합치기
tourist_profile = tourist_profile.merge(
    rating_mean,
    on="location_name",
    how="left"
)
```

```python
tourist_profile["avg_rating"] = tourist_profile["avg_rating"].round(1)
```
